```python
### FIND CORELATED MATRIX
corr_data = df.corr()
corr_data
```

### Draw heatmap to check correlation 

```python
colormap = plt.cm.viridis
plt.figure(figsize=(12,10))
sns.heatmap(corr_data,linewidths=0.1,vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True)
```

### Lets Start from target column Attration 

```python
### attration count in % 
fig_dms = (10,4)
fig, ax = plt.subplots(figsize=fig_dms)
attration = df.Attrition.value_counts()
show_in_percentage = (attration/2940 )*100
print('Attration Rate::' ,(attration/2940 )*100)
sns.barplot(y=show_in_percentage.index, x = show_in_percentage)
plt.show()
```
